```python
features = pd.read_csv("features.csv")
stores = pd.read_csv("stores.csv")
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

features.head()
```

```python
features['MarkDown1'] = features['MarkDown1'].fillna(0)
features['MarkDown2'] = features['MarkDown1'].fillna(0)
features['MarkDown3'] = features['MarkDown1'].fillna(0)
features['MarkDown4'] = features['MarkDown1'].fillna(0)
features['MarkDown5'] = features['MarkDown1'].fillna(0)

# display(features[features['MarkDown1'].notnull()].count())
# display(features[features['MarkDown2'].notnull()].count())
# display(features[features['MarkDown3'].notnull()].count())
# display(features[features['MarkDown4'].notnull()].count())
# display(features[features['MarkDown5'].notnull()].count())
```

```python
features['Date'] = pd.to_datetime(features['Date'], format="%Y-%m-%d")
train['Date'] = pd.to_datetime(train['Date'], format="%Y-%m-%d")
test['Date'] = pd.to_datetime(test['Date'], format="%Y-%m-%d")


type(features['Date'][0])
```

```python
features['CPI_lag'] = features.groupby(['Store'])['CPI'].shift(1)
features.head()
features['CPI'] = np.where(features['CPI'].isnull(), features['CPI_lag'] , features['CPI'])
features['CPI'] = np.where(features['CPI'].isnull(), 193.589304 , features['CPI'])

features['Unemployment_lag'] = features.groupby(['Store'])['Unemployment'].shift(1)
features.head()
features['Unemployment'] = np.where(features['Unemployment'].isnull(), features['Unemployment_lag'] , features['Unemployment'])
features['Unemployment'] = np.where(features['Unemployment'].isnull(), 8.335 , features['Unemployment'])

features.drop(['CPI_lag','Unemployment_lag'], axis=1, inplace=True)

features.head()
```
